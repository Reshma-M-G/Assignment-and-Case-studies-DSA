# 1. Reading dataset to Python environment




```python
#importing pandas 
import pandas as pd
#reading the excel file iris
iris_data = pd.read_excel("iris.xls")
iris_data
```

# 2. Displaying the columns

```python
#displaying only the column names of the dataset iris 
iris_data.columns
```

# 3. Calculating the mean 

```python
#importing the numpy array
import numpy as np
# calculating the mean of  each columns: sepal length,sepal width,petal length,petal width and saving to sl_mean
sl_mean=np.mean(iris_data,axis=0)
sl_mean.round(2)
```

# 4.Checking for the null values in iris

```python
#checking the null values using info and isna functions
iris_data.info()
```

# 5.Perform meaningful visualizations using the dataset


## Box plot, Swarm plot, Strip plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

###  * Box plot

```python
plt.figure(figsize=(10,10))
sns.set_theme(style="whitegrid")
plt.title("Box plotting Sepal length vs Sepal Width")
sns.boxplot(x="SL",y="SW",data=iris_data,hue="Classification",palette='Set2')
```

## * Strip plot

```python
plt.figure(figsize=(9,9))
sns.set_theme(style="whitegrid")
plt.title("Strip plot Petal Length vs Petal Width")
sns.stripplot(x="PW", y="PL",hue="Classification",linewidth=2,size=8,jitter=0.1,data=iris_data)
```

## * Violin plot

```python
plt.figure(figsize=(4,5))
sns.violinplot(x="Classification",y="PL",data=iris_data)
plt.title("Plot of PL based on classification",fontsize=16,c="Green")
plt.xlabel("Classification")
plt.ylabel("Petal Length")

```
